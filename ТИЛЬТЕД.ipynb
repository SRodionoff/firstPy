```python
from sympy import *
```

```python
a = 0.6
b = 5.5
c = 10
d = 5
x = Symbol("x")
f_x = a * (x**3) + b * (x**2) + c * x - d
```

```python
f_x
```

```python
print("Корень 1:", solve(f_x, x)[0].n())
print("Корень 2:", solve(f_x, x)[1].n())
print("Корень 3:", solve(f_x, x)[2].n())
```

```python
from sympy.plotting import plot
graph = plot(f_x,(x, 100, -100), line_color='red', title = "График функции")
```

```python
der = diff(f_x)
der
```

```python
graph2 = plot(der,(x, 100, -100), line_color='red', title = "График функции")
```

```python
graph3 = plot(f_x, der, (x, 5, -10), title = "Совмещенный график")
```

```python
solve(der < 0, x)
```

```python
solve(der > 0, x)
```

```python
der.subs(x,(-b/(2*a)))
```
